# Exercise

Circle centers: -1.5 and +1.5

Radii:.5 to 3

```python
h = [-1.5,1.5]

x = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,x)

for r in np.linspace(.5,3,15):
    for hi in h:
        Fxy = (X-hi)**2 + Y**2 - r**2
        plt.contour(X,Y,Fxy,0,colors=[r/4,r/9,r/3])

plt.axis('off')
plt.gca().set_aspect('equal')
plt.plot(h,[0,0],'k',linewidth=3)

plt.show()
```

```python
h1 = -1.5
h2 = 1.5
k = 2
r_range = np.linspace(.5,3,10)
 
x = np.linspace (-5,5,100)
X,Y = np.meshgrid(x,x)
 
for i in r_range:
    Fxy1 = (Y-k)**2+(X-h1)**2 -i**2
    plt.contour(X,Y,Fxy1,0,colors = [((i/3),(i/3),(i/3))])
 
    Fxy2 = (Y-k)**2+(X-h2)**2 -i**2
    plt.contour(X,Y,Fxy2,0,colors = [((i/3),(i/3),(i/3))])
 
plt.plot([h1,h2],[k,k],'k',linewidth = 3)
plt.gca().set_aspect('equal')
plt.axis("off")
plt.show()
```

```python
# circle parameters
a = [-5, -2.5, 2.5, 5]
 
# grid space
axlim = 10
x = np.linspace(-axlim,axlim,100)
y = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,y)
 
alpha_range = np.linspace(1, 0.5, 30)
r_range = np.linspace(.5, 3, 30)
for xi in range(0, 30):
    for ai in a:
        Fxy = (X-ai)**2 + Y**2 - (r_range[xi]*1.5)**2
        plt.contour(X,Y,Fxy,0,colors=[(r_range[xi]/3,r_range[xi]/3,0.3)], alpha=alpha_range[xi])
    
plt.gca().set_aspect('equal')
plt.plot([a[0], a[1]],[0,0],'ko-',linewidth=3)
plt.plot([a[2], a[3]],[0,0],'ko-',linewidth=3)
plt.axis('off')
plt.show()
```

```python
h = np.array([-1.5, 1.5])
k = np.array([-1.5, 1.5])
r = np.linspace(.5,3,15)
 
xaxlim = np.max((abs(r))) + np.max((abs(h)))
yaxlim = np.max((abs(r))) + np.max((abs(k)))
x = np.linspace(-xaxlim,xaxlim,100)
y = np.linspace(-yaxlim,yaxlim,100)
X,Y = np.meshgrid(x,y)
 
for ri in r:
    #h = h.tolist()
    for hi in h:
        Fxy1 = (X - hi)**2 + Y**2 - ri**2
        plt.contour(X,Y,Fxy1,colors=[(ri/3,ri/3,ri/3)])
        
for ri in r:
    #k = k.tolist()
    for ki in k:
        Fxy2 = X**2 + (Y - ki)**2 - ri**2
        plt.contour(X,Y,Fxy2,colors=[(ri/3,ri/3,ri/3)])
 
plt.gca().set_aspect('equal')
plt.plot(h,[0,0],'k',linewidth=3)
plt.plot([0,0],k,'k',linewidth=3)
plt.axis('off')
plt.show()
```
