# Exercise

Vertex:(h,k)

Focus:(h,k + p)

Directrix: y = k - p

```python
#parameters
a =1  #1/4p = a, 1=a4p,1/4p
p = 1/(4*a)
h = 1
k = -2
n = 200

#x-axis point to evaluate the function
x = np.linspace(h-2,h+2,n)

#create the function
y = a*(x-h)**2 + k

#plot the parabola
plt.plot(x,y,label='Parabola')

#plot the vertex
plt.plot(h,k,'ro',label='Vertex')

#plot the focus
plt.plot(h,k+p,'go',label='Focus')

#plot the directrix
d = k-p
plt.plot(x[[0,-1]],[d,d],label='Directrix')

plt.grid()
plt.legend()
plt.show()
```
