```python
noOfEllipse = 106
noOfSteps = 100
 
k = np.linspace(-9, 9, noOfEllipse)
h = np.ones(len(k))
a = abs(np.array(k))
b = np.ones(len(k))
 
plt.figure(figsize=(12, 7))
 
for i in range(noOfEllipse):
 
    xlim = a[i] + abs(h[i])
    ylim = b[i] + abs(k[i])
    x = np.linspace(-xlim, xlim, noOfSteps)
    y = np.linspace(-ylim, ylim , noOfSteps)
    X,Y = np.meshgrid(x, y)
 
    ellipse = (X - h[i])**2/a[i]**2 + (Y - k[i])**2/b[i]**2 - 1/(i+1)
 
    plt.contour(X, Y, ellipse, 0, colors=[np.clip(i/noOfEllipse, 0.1, 0.4), np.clip(i/noOfEllipse, 0.1, 0.5), np.clip(i/noOfEllipse, 0.5, 1)])
 
plt.axis('off')
plt.gca().set_aspect('equal')
plt.show()
```
