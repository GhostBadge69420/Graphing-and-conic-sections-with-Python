```python
# Number of hyperbolas per axis
n = 5
# Changing 'a' and 'b' values
ab = np.linspace(1,5,n)
# Shown area
axlim = 40; 
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)
 
# Prepare 'star' rotations
# Star distances
A  = 25
# Number of stars + 1 
pn = 7 
phi = np.linspace(0,2*np.pi,pn)
 
 
# Draw each star (hyperbolas) within a circle
for j in range(0,pn):
    h = A*np.cos(phi[j])
    k = A*np.sin(phi[j])   
    # Color for the wheele spokes
    c = j/pn
    
    # Create Hyperbolas
    for i in range(0,n):
        F1 = (X-h)**2/ab[i]**2 - (Y-k)**2/ab[i]**2 - 1 
        # If the axis are turned off, we don't need the 'X,Y' in front of 'F1'.
        # But, you need to define the 'X,Y' coordinates and draw the contours ...
        # ... in coordinates instead of indices to use it together with 'plt.plot'.
        plt.contour(X,Y,F1,0,colors = [i/n,i/n,0]) 
        # Swap the 'X' and 'Y' by changing the operator '+' and '-'
        F2 = -(X-h)**2/ab[i]**2 + (Y-k)**2/ab[i]**2 - 1 
        plt.contour(X,Y,F2,0,colors = [i/n,i/n,i/n]) 
        # Draw each star center
        F3 = (X-h)**2 + (Y-k)**2 - 4
        plt.contour(X,Y,F3,0,colors = 'r') 
        
        # Draw each wheel spoke and be sure the plt.contour(X,Y, ..) are coordinates. 
        plt.plot([0,h],[0,k],'-.',color=[c*0.8,c*0.9,0],linewidth='2' )
 
# Draw the center of the circle        
P1 = X**2 + Y**2 - A
plt.contour(X,Y,P1,0,colors = 'k') 
 
# Graphing stuff
# plt.title('Rotating Stars',loc = 'right')
plt.axis('off')
plt.gca().set_aspect('equal')
 
# Export PDF
plt.savefig('rotating-Stars-1.pdf')
plt.savefig('rotating-Stars-1.jpeg')
plt.show()
```
